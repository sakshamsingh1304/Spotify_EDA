	
### About the Dataset:  ###
This dataset contains a comprehensive list of the most famous songs of 2023 as listed on Spotify. The dataset offers a wealth of features beyond what is typically available in similar datasets. It provides insights into each song's attributes, popularity, and presence on various music platforms. The dataset includes information such as track name, artist(s) name, release date, Spotify playlists and charts, streaming statistics, Apple Music presence, Deezer presence, Shazam charts, and various audio features.



```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

df.head()
```

### Key Features: ###
track_name: Name of the song                                                                                                                            
artist(s)_name: Name of the artist(s) of the song                                                                                                       
artist_count: Number of artists contributing to the song                                                                                               
released_year: Year when the song was released                                                                                                        
released_month: Month when the song was released                                                                                                      
released_day: Day of the month when the song was released                                                                                              
in_spotify_playlists: Number of Spotify playlists the song is included in                                                                             
in_spotify_charts: Presence and rank of the song on Spotify charts                                                                                    
streams: Total number of streams on Spotify                                                                                                             
in_apple_playlists: Number of Apple Music playlists the song is included in                                                                             
in_apple_charts: Presence and rank of the song on Apple Music charts                                                                                   
in_deezer_playlists: Number of Deezer playlists the song is included in                                                                                 
in_deezer_charts: Presence and rank of the song on Deezer charts                                                                                       
in_shazam_charts: Presence and rank of the song on Shazam charts                                                                                       
bpm: Beats per minute, a measure of song tempo                                                                                                         
key: Key of the song                                                                                                                                   
mode: Mode of the song (major or minor)                                                                                                                 
danceability_%: Percentage indicating how suitable the song is for dancing                                                                              
valence_%: Positivity of the song's musical content                                                                                                    
energy_%: Perceived energy level of the song                                                                                                           
acousticness_%: Amount of acoustic sound in the song                                                                                                   
instrumentalness_%: Amount of instrumental content in the song                                                                                         
liveness_%: Presence of live performance elements                                                                                                      
speechiness_%: Amount of spoken words in the song

```python
#removing non-numeric row in stream
to_drop = df.loc[df.streams.str.isnumeric() == False]['streams'].index
df.drop(to_drop, inplace=True)
```

```python
#removing commas(string to numerical)
transform_cols = ["in_deezer_playlists", "in_shazam_charts"]
for col in transform_cols:
    df[col] = df[col].replace(',','', regex=True)
```

```python
#convert columns to numerical feature
convert_cols = ["streams", "in_deezer_playlists", "in_shazam_charts"]
for col in convert_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

**Average Streams Per Mode**


Objective: Analyze the average number of streams for tracks in Major vs. Minor modes to see how the mode impacts popularity.

```python
# Average streams for Major and Minor mode tracks
average_streams_per_mode = df.groupby("mode")["streams"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=['Major', 'Minor'], y=average_streams_per_mode.values, palette='coolwarm')
plt.title('Average Streams per Mode', fontsize=14)
plt.ylabel('Average Streams')
plt.xlabel('Mode')
plt.show()
```

**Energy Distribution of Tracks Across Modes**


We can visualize the distribution of energy% across major and minor modes using a violin plot.

**Monthly Streams Trend**


Objective: Understand how streaming numbers vary across months by visualizing the cumulative streams for all tracks released in each month.

```python
# Monthly streams trend
monthly_streams = df.groupby("released_month")["streams"].sum().reset_index()
monthly_streams["released_month"] = pd.Categorical(monthly_streams["released_month"], categories=range(1, 13), ordered=True)
monthly_streams = monthly_streams.sort_values("released_month")

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_streams, x="released_month", y="streams", marker="o", color="teal")
plt.title("Monthly Streaming Trends in 2023", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Cumulative Streams")
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(visible=True)
plt.show()
```

**Findings**

1. **Most Streamed Track and Artist:**  
   - The track with the highest number of streams in 2023 is **Blinding Lights** by **The Weeknd**.  
   - **The Weeknd** also leads as the artist with the most cumulative streams across all tracks, highlighting his dominance on Spotify.

2. **Track Release Trends:**  
   - Most top tracks were released during **January, May, and June**, reflecting strategic timing for music releases.
   - The months **February** and **December** saw fewer popular song releases, potentially due to seasonal trends in the music industry.

3. **Track Modes:**  
   - The majority of the top tracks are in the **Major mode**, known for producing lighter, brighter, and more cheerful sounds.  
   - Tracks in the **Minor mode**, though fewer, may cater to niche preferences.

4. **Energy Levels Across Release Months (Box Plot):**  
   - **June and August** feature tracks with higher energy levels, aligning with summer playlists that often favor vibrant, upbeat tracks.
   - Tracks released in **November** showed the lowest energy levels, possibly catering to calmer, end-of-year moods.

5. **Streams vs. Acousticness (Scatter Plot):**  
   - A negative correlation was observed: Tracks with **higher acousticness** tend to have **lower stream counts**, indicating that heavily acoustic songs are less mainstream or appeal to niche audiences.  
   - The most-streamed tracks tend to feature moderate to low acousticness, balancing production elements.

6. **Influence of Playlists on Streams:**  
   - Tracks included in more **Spotify playlists** and **charts** have a significantly higher likelihood of accumulating large streams, emphasizing the platform's influence on a track's success.

7. **Energy and Streams Correlation:**  
   - High-energy tracks generally receive more streams, suggesting that energetic music resonates more with Spotify's active audience.

8. **Monthly Streaming Trends:**  
   - Streams peak during **summer months** (June-August), coinciding with the release of high-energy tracks and increased user activity.

9. **Acousticness and Streams Relationship:**  
   - While heavily acoustic tracks attract fewer streams, those with balanced production, blending acoustic and digital elements, tend to perform well.


The dataset analysis offers valuable insights into the preferences and trends shaping the music industry in 2023. Tracks released strategically during specific months, featuring vibrant energy and moderate acousticness, and promoted effectively on Spotify playlists are more likely to become chart-topping hits. **The Weeknd** continues to be a prominent artist, both in terms of streams and influence on global music trends.
